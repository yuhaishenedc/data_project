## 1 项目背景

近些年来，心血管疾病引发的死亡已经持续成为居民死亡的首要原因。根据世界卫生组织的调查，全球共有 1700 万人死于心脏病发作和中风。很多心脏病患者死亡的原因是工作超负荷、精神压力过大等问题。而且心脏病的患病人群也越来越年轻化。因此，心脏病的诊断与治疗已经成为医学研究领域中的重中之重。但心脏病的种类繁多，影响因素多样，诊断心脏病是一项复杂而重要的任务，如何提高心脏病的诊断效率已经成为急需解决的问题。本次实验选题针对心脏病的诊断，期望通过各种机器学习算法给出一个能够有效判断心脏病的系统。

以下是数据集中每个属性的含义说明：

* Age(年龄)：年龄是患心血管疾病或心脏病的最重要风险因素，每十年风险大约增加三倍。冠状动脉脂肪条纹可以在青春期开始形成。据估计，死于冠心病的人中有 82% 年龄在 65 岁及以上。同时，55 岁后中风的风险每十年翻一番。
* Sex(性别)：男性比绝经前女性患心脏病的风险更高。一旦过了更年期，有人认为女性的风险与男性相似，尽管来自世界卫生组织和联合国的最新数据对此提出了质疑。如果女性患有糖尿病，她比患有糖尿病的男性更容易患心脏病。
* Chest Pain(胸痛类型)：心绞痛是当您的心肌没有获得足够的富氧血液时引起的胸痛或不适。它可能会感觉像是在你的胸部有压力或挤压。不适也可能发生在您的肩膀、手臂、颈部、下巴或背部。心绞痛甚至可能感觉像是消化不良。
* Resting Blood Pressure(静止血压)：随着时间的推移，高血压会损害为心脏供血的动脉。与肥胖、高胆固醇或糖尿病等其他疾病一起发生的高血压会进一步增加您的风险。
* Serum Cholesterol(血清胆固醇含量)：高水平的低密度脂蛋白 (LDL) 胆固醇（“坏”胆固醇）最有可能使动脉变窄。高水平的甘油三酯（一种与饮食相关的血脂）也会增加心脏病发作的风险。然而，高水平的高密度脂蛋白 (HDL) 胆固醇（“好”胆固醇）会降低心脏病发作的风险。
* Fasting Blood Sugar (空腹血糖浓度)：胰腺分泌的激素（胰岛素）不足或对胰岛素没有正确反应会导致身体血糖水平升高，从而增加心脏病发作的风险。
* Resting ECG (平静状态下的心电图)：对于心血管疾病风险较低的人群，USPSTF 得出的结论是，通过静息或运动心电图进行筛查的潜在危害等于或超过潜在益处。对于中高风险人群，目前的证据不足以评估筛查的利弊平衡。
* Max heart rate achieved (最大心率)：与心率加快相关的心血管风险增加与高血压所观察到的风险增加相当。研究表明，心率每分钟增加 10 次与心源性死亡风险增加至少 20% 相关，这种风险增加与收缩血增加所观察到的风险增加相似10 毫米汞柱的压力。
* Exercise induced angina (运动诱发的心绞痛)：与心绞痛相关的疼痛或不适通常感觉紧绷、抓握或挤压，并且可以从轻微到严重不等。心绞痛通常在您的胸部中央感觉到，但可能会扩散到您的肩膀或背部，颈部，下巴或手臂。它甚至可以在你的手中感觉到。 o 心绞痛的类型稳定型心绞痛/心绞痛 b.不稳定型心绞痛 C.变体（Prinzmetal）心绞痛 d。微血管心绞痛。
* Peak exercise ST segment (峰值运动 ST 段)：当 J 点后 60-80 ms 出现水平或向下倾斜的 ST 段压低 ≥ 1 mm 时，认为跑步机心电图压力测试异常。具有上斜 ST 段压低的运动心电图通常被报告为“模棱两可”的测试。一般而言，在较低的工作量（以 MET 计算）或心率发生水平或向下倾斜的 ST 段压低表明预后较差，多血管疾病的可能性较高。 ST 段压低的持续时间也很重要，因为峰值压力后恢复的时间延长与跑步机心电图压力测试的阳性结果一致。另一个高度提示严重 CAD 的发现是 ST 段抬高 > 1 mm（通常提示透壁缺血）；这些患者经常被紧急转诊进行冠状动脉造影。

## 3 实验目标

本次实验主要目的是对心脏病数据集，进行数据分析，利用历史心脏病数据对心脏病进行诊断。该数据集包含 76 个属性，我们的实验中使用了其中 13个属性的子集，诸如年龄、性别、胸痛等。通过使用数据挖掘相关技术，如K近邻、神经网络、决策树、随机森林等。

* K近邻分类：在特征空间中，如果一个样本附近的k个最近(即特征空间中最邻近)样本的大多数属于某一个类别，则该样本也属于这个类别。
* LinearSVC：线性支持分类，适应更大的数据集。
* 随机森林分类：一种监督学习算法，随机森林算法在数据样本上创建决策树，然后从每个样本中获取预测，最后通过投票选择最佳解决方案。是一种集成方法，通过对结果求平均值来减少过度拟合。
* LightGBM：在传统的GBDT算法进行了优化的算法。采用基于Histogram的决策树算法，。使用单边梯度采样（GOSS）来减少大量只具有小梯度的数据实例。
* MLP：基于生物神经元模型得到多层感知机的基本结构，第一层为输入层，最后一层为输出层，中间层为隐层。本实验采用4层结构。MLP神经网络不同层之间是全连接的。

## 6 构建baseline判定模型，观察结果

本次项目主要使用以下三个评价指标：

* Precision

$P=\frac{TP}{TP+FP}$

* Recall

$R=\frac{TP}{TP+FN}$

* Accuracy

$A=\frac{TP+TN}{TP+FP+TN+FN}$


其中：

* TP（true positive）：在判定为positive的样本中，判断正确的数目
* FP（false positive）：在判定为positive的样本中，判断错误的数目
* TN（true negative）：在判定为negative的样本中，判断正确的数目
* FN（false negative）：在判定为negative的样本中，判断错误的数目

#### 6.1 决策树

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier(criterion='entropy')

accuracy=0
precision=0
recall=0
for train,test in kf.split(dataframe):
    train=dataframe.iloc[train]
    test=dataframe.iloc[test]
    train_x=train.drop(columns=['output'])
    train_y=train['output']
    test_x=test.drop(columns=['output'])
    test_y=pd.DataFrame(test['output'])
    clf.fit(train_x,train_y)
    result=clf.predict(test_x)
    accuracy+=accuracy_score(test_y,result)
    precision+=precision_score(test_y,result)
    recall+=recall_score(test_y,result)

print('precision is:',precision/10,' recal is:',recall/10, ' accuracy is:',accuracy/10)
```

#### 6.2 K近邻

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)

accuracy=0
precision=0
recall=0
for train,test in kf.split(dataframe):
    train=dataframe.iloc[train]
    test=dataframe.iloc[test]
    train_x=train.drop(columns=['output'])
    train_y=train['output']
    test_x=test.drop(columns=['output'])
    test_y=pd.DataFrame(test['output'])
    knn.fit(train_x,train_y)
    result=knn.predict(test_x)
    accuracy+=accuracy_score(test_y,result)
    precision+=precision_score(test_y,result)
    recall+=recall_score(test_y,result)

print('precision is:',precision/10,' recal is:',recall/10, ' accuracy is:',accuracy/10)
```

#### 利用MLP进行分类

```python
import torch
from torch import nn


class MLP(nn.Module):
    def __init__(self):
        super(MLP,self).__init__()
        
        self.fc1=nn.Sequential(
            nn.Linear(11,64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.1),
        )
        
        self.fc2=nn.Sequential(
            nn.Linear(64,128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),
        )
        
        self.fc3=nn.Sequential(
            nn.Linear(128,64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.5),
        )
        
        
        
        self.out=nn.Sequential(
            nn.Linear(64,2),
        )
    
    def forward(self,x):
        x=self.fc1(x)
        x=self.fc2(x)
        x=self.fc3(x)
        output=self.out(x)
        return output

net=MLP()
```

```python
import torch.utils.data.dataset as Dataset
import torch.utils.data as data
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader, TensorDataset

def get_dataLoader(df,batch_size):
    y=df['target']
    x=df.drop(columns=['target'])
    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.1, random_state=2018)
    train_set=TensorDataset(torch.from_numpy(x_train.values).to(torch.float32),torch.from_numpy(y_train.values).to(torch.float32))
    test_set = TensorDataset(torch.from_numpy(x_val.values).to(torch.float32),torch.from_numpy( y_val.values).to(torch.float32))
    train_loader = DataLoader( train_set,batch_size=batch_size, shuffle=True,drop_last=True)
    test_loader = DataLoader(test_set, batch_size=batch_size,drop_last=True)
    return train_loader,test_loader

train_loader,test_loader=get_dataLoader(df,64)
```

```python
epochs = 500
optimizer = torch.optim.Adam(net.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

net = net.to(device)
loss_fn = loss_fn.to(device)

train_loss_list=[]
test_loss_list=[]
train_accuracy_list=[]
test_accuracy_list=[]

def train(net, train_loader, optimizer, device, loss_fn):
    net.train()
    train_loss = 0
    corrects=0
    length=0
    for data in train_loader:
        
        optimizer.zero_grad()
        x, targets = data
        length+=len(targets)
        x = x.to(device)
        targets = targets.to(device)
        y_pred = net(x)
        y_hat = torch.argmax(y_pred, 1)
        corrects+=(y_hat==targets).sum()
        
        loss = loss_fn(y_pred, targets.long())
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        
    train_loss_list.append(train_loss)
    train_accuracy_list.append(corrects/length)
    return train_loss

def eval(net, test_loader, device, loss_fn):
    net.eval()
    test_loss = 0
    corrects=0
    length=0
    for data in test_loader:
        X, targets = data
        length+=len(targets)
        X = X.to(device)
        targets = targets.to(device)
        y_pred = net(X)
        y_hat = torch.argmax(y_pred, 1)
        loss = loss_fn(y_pred, targets.long())
        test_loss += loss.item()
        corrects+=(y_hat==targets).sum()
    test_loss_list.append(test_loss)
    test_accuracy_list.append(corrects/length)
    return test_loss

for epoch in range(epochs):
    train_loss=train(net, train_loader, optimizer, device, loss_fn)
    loss=eval(net, test_loader, device, loss_fn)
```

```python
def draw_loss(loss1,loss2):
    plt.figure(figsize=(16,6))
    plt.cla()
    x=range(1,len(loss1)+1)
    y1=loss1
    y2=loss2
    plt.title("loss during train",fontsize=10)
    plt.plot(x,y1,'.-',label='train_loss')
    plt.plot(x,y2,'.-',label='test_loss')
    plt.ylabel('loss value',fontsize=10)
    plt.xlabel('epoch',fontsize=10)
    plt.legend()
    plt.show()

draw_loss(train_loss_list,test_loss_list)
```

```python
def draw_accuracy(accuracy1,accuracy2):
    plt.figure(figsize=(16,6))
    plt.cla()
    x=range(1,len(accuracy1)+1)
    y1=accuracy1
    y2=accuracy2
    plt.title("accuracy during training",fontsize=10)
    plt.plot(x,y1,'.-',label='train_accuracy')
    plt.plot(x,y2,'.-',label='test_accuracy')
    plt.ylabel('accuracy value',fontsize=10)
    plt.xlabel('epoch',fontsize=10)
    plt.legend()
    plt.show()
    
draw_accuracy(train_accuracy_list,test_accuracy_list)
```
